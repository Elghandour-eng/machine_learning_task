## LR Full Implementation

### Step1: Initialize parameters (theta_0 & theta_1) with random value or simply zero. Also choose the Learning rate.

```python
#Step1: Initialize parameters (theta_0 & theta_1) with random value or simply zero. Also choose the Learning rate.
theta_0 = 0 
theta_1=0
alpha = 0.0001
```

### Step2: Use (theta_0 & theta_1) to predict the output h(x)= theta_0 + theta_1 * x.
#### Note: you will need to iterate through all data points

```python
# Step2: Use (theta_0 & theta_1) to predict the output h(x)= theta_0 + theta_1 * x.
h_x=np.zeros(len(X))
for i in range(len(X)):
    h_x[i] = theta_0 + theta_1 * X[i]
print(f"h(x) :",h_x)
```

### Step3: Calculate the MSE Cost function ùë±(theta_0,theta_1 ).
<b>Note:</b> You better use either dot product or norm square of the error vector.

```python
# Step3: Calculate the cost function J(theta_0,theta_1) = 1/2m * sum(h(x)-y)^2
# calculate the Error Vector with dot product
error_vector=(h_x-y)
print("Error Vector :\n",error_vector)

cost = 1/(2*len(X)) * np.sum((h_x-y)**2)
print("J =",cost)
```

### Step4: Calculate the gradient components for theta_0 and theta_1.
<b>Note:</b> You can use the error vector calculated in the previous step.

```python
# Step4: Calculate the gradient components dJ/dtheta_0 and dJ/dtheta_1
dj_dtheta_0 = 1/len(X) * np.sum(h_x-y)
dj_dtheta_1 = 1/len(X) * np.sum((h_x-y)*X)
print("dJ/dtheta_0 =",dj_dtheta_0)
print("dJ/dtheta_1 =",dj_dtheta_1)
```

### Step5: Update the parameters (simultaneously).

```python
# Step5: Update the parameters theta_0 and theta_1
theta_0_new=theta_0-alpha*dj_dtheta_0
theta_1_new=theta_1-alpha*dj_dtheta_1

print("theta_0_new =",theta_0_new)
print("theta_1_new =",theta_1_new)
```
